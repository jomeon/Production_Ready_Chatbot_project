# APIS AI - Testy manualne Chatbota 
Poniższy notatnik prezentuje działanie głównej logiki chatbota (`APISChatbot`) połączonej z lokalnym modelem Ollama. 

**Wymagania wstępne do uruchomienia:**
1. Aktywne środowisko: `conda activate apis_chatbot`
2. Uruchomiony w tle silnik Ollama z modelem `llama3` (`ollama run llama3`).

```python
import sys
import os

sys.path.append(os.path.abspath('..'))

from app.chatbot_core import APISChatbot

print("Inicjalizacja chatbota...")
bot = APISChatbot()
print(f"Chatbot gotowy. Model: {bot.model_name}")
print(f"Bieżąca historia: {bot.history}")
```

```python
pytanie_1 = "Cześć, wymień 3 najpopularniejsze języki programowania."
print(f"Użytkownik: {pytanie_1}")

odpowiedz_1 = bot.generate_response(pytanie_1, temperature=0.5)
print(f"\nAPIS Bot:\n{odpowiedz_1}")
```

```python
pytanie_2 = "Z tych trzech, który jest najlepszy do analizy danych?"
print(f"Użytkownik: {pytanie_2}")

odpowiedz_2 = bot.generate_response(pytanie_2, temperature=0.7)
print(f"\nAPIS Bot:\n{odpowiedz_2}")
```
